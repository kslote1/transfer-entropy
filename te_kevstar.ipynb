```python
import numpy as np, sys, random
def toseconds(x):
    return 0 #placeholder
timeseries=np.zeros([365,4])
timeseries[:,2]=np.loadtxt('NICS_daily.np')
ts_SP=np.loadtxt('ts_AR_SP.np')
ix1=ts_SP[:,0].astype(int)
timeseries[ix1,0]=ts_SP[:,1]
ts_FoR=np.loadtxt('ts_AR_FoR.np')
ix2=ts_FoR[:,0].astype(int)
timeseries[ix2,1]=ts_FoR[:,1]
ts_AR=np.loadtxt('ts_AR.np')
ix3=ts_AR[:,0].astype(int)
timeseries[ix3,3]=ts_AR[:,1]
timeseries[:,0]/=timeseries[:,3]
timeseries[:,1]/=timeseries[:,3]
mn=170
mx=250
timeseries=timeseries[mn:mx,:]
def binary_conditional_MI(x,y,z):
    x=x.astype(np.int32)
    y=y.astype(np.int32)
    z=z.astype(np.int32)
    xyz=x*4+y*2+z
    pXYZ=np.zeros(8)
    for s in range(8):
        pXYZ[s]=np.sum(np.equal(xyz,s))/len(xyz)
    z=np.mod(xyz,2)
    yz=np.mod(xyz,4)
    xz=np.bitwise_xor(xyz,2)
    pZ=np.bincount(z)/len(z)
    pXZ=np.bincount(xz)/len(xz)
    pYZ=np.bincount(yz)/len(yz)
    TE=0
    for s in range(8):
        if pXYZ[s]>0:
            TE+=pXYZ[s]*np.log(pZ[s%2]*pXYZ[s]/(pXZ[s^2]*pYZ[s%4]))
    return TE

i=1
def quantize(x):
    return np.diff(x)>0
Y1=quantize(timeseries[1:,2])
Y0=quantize(timeseries[:-1,2])
X0=quantize(timeseries[:-1,i])
def shuffled(x):
    return quantize(np.random.permutation(x))

te=binary_conditional_MI(Y1,X0,Y0)
print(te)

Ntest=10000
Stest=np.zeros(Ntest)
for t in range(Ntest):
    Stest[t]=binary_conditional_MI(Y1,shuffled(timeseries[:-1,i]),Y0)
print((Stest)[int(Ntest*0.95)], Stest.min(),Stest.max())
```
