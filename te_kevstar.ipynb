```python
import numpy as np, sys, random
def toseconds(x):
    return 0 #placeholder
timeseries=np.zeros([365,4])
timeseries[:,2]=np.loadtxt('NICS_daily.np')
ts_SP=np.loadtxt('ts_AR_SP.np')
ix1=ts_SP[:,0].astype(int)
timeseries[ix1,0]=ts_SP[:,1]
ts_FoR=np.loadtxt('ts_AR_FoR.np')
ix2=ts_FoR[:,0].astype(int)
timeseries[ix2,1]=ts_FoR[:,1]
ts_AR=np.loadtxt('ts_AR.np')
ix3=ts_AR[:,0].astype(int)
timeseries[ix3,3]=ts_AR[:,1]
timeseries[:,0]=np.nan_to_num(timeseries[:,0]/timeseries[:,3])
timeseries[:,1]=np.nan_to_num(timeseries[:,1]/timeseries[:,3])
mn=170
mx=250
timeseries=timeseries[mn:mx,:]
def te(x,y,y1):
    x=x.astype(np.int32)
    y=y.astype(np.int32)
    y1=y1.astype(np.int32)
    p_joint=np.zeros(8)
    p_num=np.zeros(8)
    p_denom=np.zeros(8)
    s=0
    for i0 in range(2):
        for i1 in range(2):
            for j0 in range(2):
                n_y1yx=0
                n_y1y=0
                N_joint=0
                N_num=0
                N_denom=0
                for n in range(len(x)):
                    n_y1yx+=1 if (x[n]==j0 and y[n]==i0 and y1[n]==i1) else 0
                    n_y1y+=1 if y[n]==i0 and y1[n]==i1 else 0
                    N_joint+=1
                    N_num+=1 if x[n]==j0 and y[n]==i0 else 0
                    N_denom+=1 if y[n]==i0 else 0
                p_joint[s]=n_y1yx/N_joint if N_joint>0 else 0
                p_num[s]=n_y1yx/N_num if N_num>0 else 0
                p_denom[s]=n_y1y/N_denom if N_denom>0 else 0
                s+=1   
        
    TE=0
    for s in range(8):
        if p_joint[s]>0:
            TE+=p_joint[s]*np.log(p_num[s]/p_denom[s])
    return TE

i=1
def quantize(x):
    return 1*(x>np.median(x))
Y1=quantize(timeseries[1:,2])
Y0=quantize(timeseries[:-1,2])
X0=quantize(timeseries[:-1,i])
def shuffled(x):
    return quantize(np.random.permutation(x))

t=te(X0,Y0,Y1)
print(t)
Ntest=10000
Stest=np.zeros(Ntest)
for T in range(Ntest):
    Stest[T]=te(shuffled(timeseries[:-1,i]),Y0,Y1)
print((Stest)[int(Ntest*0.95)], Stest.min(),Stest.max())
print(1-len(np.where(Stest<t)[0])/Ntest)
```
